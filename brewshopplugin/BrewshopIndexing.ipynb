```python
# split string on every '/' (ignore http://)
import time 

# time in YYYY-MM-DD HH:MM:SS format
timestamp = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())


def process_url(url):
    stringlist = url[8:].split('/')
    if len(stringlist) == 2:
        object = {
            'updated': timestamp,
            'url': url,
            'domain': stringlist[0],
            'category': stringlist[1],
        }
    elif len(stringlist) == 3:
        object = {
            'updated': timestamp,
            'url': url,
            'domain': stringlist[0],
            'category': stringlist[1],
            'subcategory': stringlist[2]
        }
    elif len(stringlist) == 4:
        object = {
            'updated': timestamp,
            'url': url,
            'domain': stringlist[0],
            'category': stringlist[1],
            'subcategory': stringlist[2],
            'product': stringlist[3]
        }
    elif len(stringlist) == 5:
        object = {
            'updated': timestamp,
            'url': url,
            'domain': stringlist[0],
            'category': stringlist[1],
            'subcategory': stringlist[2],
            'subcategory2': stringlist[3],
            'product': stringlist[4]
        }
    elif len(stringlist) == 6:
        object = {
            'updated': timestamp,
            'url': url,
            'domain': stringlist[0],
            'category': stringlist[1],
            'subcategory': stringlist[2],
            'subcategory2': stringlist[3],
            'subcategory3': stringlist[4],
            'product': stringlist[5]
        }
    elif len(stringlist) == 7:
        object = {
            'updated': timestamp,
            'url': url,
            'domain': stringlist[0],
            'category': stringlist[1],
            'subcategory': stringlist[2],
            'subcategory2': stringlist[3],
            'subcategory3': stringlist[4],
            'subcategory4': stringlist[5],
            'product': stringlist[6]
            
        }

    return object
    
```

```python
def process_text(dictionary):
    product_tiltle = dictionary['product_title']
    product_ingress = dictionary['product_ingress']
    product_description = dictionary['product_description']
    product_price = dictionary['product_price']
    link = dictionary['url']
    text = f"Produkttittel: {product_tiltle} \n Produkt ingress: {product_ingress} \n Produktbeskrivelse: {product_description} \n Pris: {product_price} \n link: {link}"
    return text
```

```python
print(process_text(data[0]))
```

```python
documents = []
for i in data:
    page_content = process_text(i)
    metadata = process_url(i['url'])
    if metadata == None:
        metadata = {'error': 'No metadata found'}
        print("No metadata found")
        print(i)
    document = Document(page_content=page_content, metadata=metadata)
    documents.append(document)
```

```python
# create vector database
embeddings = OpenAIEmbeddings()
db = Chroma.from_documents(documents, embeddings, persist_directory="./brewshop_chroma_db")

# persist vector database
db.persist()
print("Vector database persisted.")
```
